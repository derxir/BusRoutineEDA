```python
import pandas as pd
from datetime import date, timedelta
```

```python
colnames=['date','week','0640','0655','0710','0725','0740','0755','0810','0825','0845','0905','0906','1705','1725','1745','1805','1825','1845','1905','1920','1935','1950','2005']
weeks = ['Monday', 'Tuesday', 'Wednesday','Thursday','Friday']
```

```python
def insert_template(template, df, first_date):
    r, _ = template.shape
    for w in range(len(weeks)):
        template.loc[r] = [first_date + timedelta(days=w), weeks[w]] + [df.iloc[i,j] for i in [1+w,8+w] for j in range(1,12)]
        r+=1
```

```python
def extract(filepath, start_date):
    print('\nprocessing year ' + str(start_date.year) + '...')
    template = pd.DataFrame(columns=colnames)
    print('week',end =" ")
    for weekNo in range(52):
        print(str(weekNo), end =" ")
        insert_template(template, pd.read_excel(filepath, sheetname=weekNo), start_date)
        start_date += timedelta(days=7)
    del template['0906'] # drop dummy column
    return template
```

```python
d16 = extract('data/2016.xlsx', date(2015,12,29))
d17 = extract('data/2017.xlsx', date(2017,1,2))
d18 = extract('data/2018.xlsx', date(2018,1,2))
#d19 = extract('data/2019.xlsx', date(2019,1,2))
#d20 = extract('data/2020.xlsx', date(2020,1,2))
```
